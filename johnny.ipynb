```python
import pandas as pd

df = pd.read_csv('leak_reintro_raw_data.txt', delimiter='\t')
lots = df.LOT_ID.unique()
recs = []

for lot in lots:
    lot_df = df[df.LOT_ID == lot].sort_values(by='HIST_SEQ').copy()
    pivoted_dict = {}
    pivoted_dict['Casting ID'] = lot
    chamber_seq = 1
    for i, rec in enumerate(lot_df.values):
        if i == 0: 
            if rec[2] == 'START': 
                continue
            else: 
                pivoted_dict['LEAK_CHAMBER' + str(chamber_seq)] = rec[4]
                chamber_seq += 1
        else:
            if rec[2] == 'START': 
                pivoted_dict['LEAK_CHAMBER' + str(chamber_seq)] = rec[4]
                pivoted_dict['RESULT' + str(chamber_seq)] = 'PASS'
            else:
                pivoted_dict['LEAK_CHAMBER' + str(chamber_seq)] = rec[4]
                pivoted_dict['RESULT' + str(chamber_seq)] = 'FAIL'
                chamber_seq += 1
    if chamber_seq < 4:
        for csq in range( chamber_seq+1, 5 ):
            pivoted_dict['LEAK_CHAMBER' + str(csq)] = ''
            pivoted_dict['RESULT' + str(csq)] = ''
    recs.append(pivoted_dict)

pdf = pd.DataFrame(recs)
pdf.fillna('')
pdf.to_csv('reintro_data.csv')
```
